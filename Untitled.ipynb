```python
import pandas as pd

#usecols=['Image Number','Scientific Name']

df=pd.read_excel('C://Users//Sohan-PC//Documents//Shared//Insect//detail.xls',usecols=['Image Number','Scientific Name'],converters={'Image Number':str,'Scientific Name':str})
df.head()
```

```python
st=[]
for x in df['Scientific Name']:
    st.append(x)
st=set(st)
print(len(st))
```

```python
path='C://Users//Sohan-PC//Documents//Shared//Insect//'
list1=[]
list2=[]
D={}
for x in df['Image Number']:
    y=str(x)
    y+='.jpg'
    list1.append(y)
for x in df['Scientific Name']:
    list2.append(x)
print(len(list1),len(list2))
for i in range(len(list1)):
    D[list1[i]]=list2[i]
```

```python
import os

for x in st:
    new_dir=path+x
    try:
        os.mkdir(new_dir)
    except:
        break
```

```python
import shutil

new_path=path+'image//'
print(new_path)
```

```python
for file in os.listdir(new_path):
    x=D[file]
    dest=path+x+'//'
    source=new_path+file
    print(dest,source)
    shutil.move(source,dest)
```
